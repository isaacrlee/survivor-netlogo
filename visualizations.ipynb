```python
import pandas as pd
```

```python
def output_challenge_eliminated_df():
    with open("challenge-eliminated-list.txt") as f:
        content = f.readlines()
    content = [x.strip().split(',') for x in content]
    
    # Clean Text
    for i in range(10):
        content[i][0] = 'Tribe ' + content[i][0]
    for i in range(len(content)):
        content[i][1] = content[i][1].title()[1:len(content[i][1]) - 1]
    for i in range(10,18):
        content[i][0] = content[i][0].title()[1:len(content[i][0]) - 1]

    df = pd.DataFrame(content, columns=['Challenge Winner', 'Eliminated Contestant'])
    return df

def output_contestant_resumes_df():
    with open("contestant-resumes-list.txt") as f:
        content = f.readlines()
    content
    content = [x.strip().split(',') for x in content]

    df = pd.DataFrame(content, columns=['Contestant', 'Tribe', 'Finish', '# Individual Challenges Won', 'Physical'])
    return df

def output_voting_histories_df():
    with open("voting-histories-list.txt") as f:
        content = f.readlines()
    content = [x.strip().split(',') for x in content]
    target_length = len(content[0])
    rows = []

    # pad rows
    for i in range(len(content)):
        length = len(content[i])
        content[i].extend([""] * (target_length - length))
        rows.append(content[i].pop(0))

    for i in range(len(content)):
        for j in range(len(content[0])):
            if content[i][j] == 'nobody':
                content[i][j] = '-'
    df = pd.DataFrame(content, index=rows, columns=rows[:1:-1])
    return df
```

```python
output_challenge_eliminated_df()
```

```python
output_contestant_resumes_df()
```
